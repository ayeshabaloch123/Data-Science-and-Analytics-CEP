##Feature Engineering

Handling Missing Data


```python
data.fillna(data.mean(), inplace=True)
print("\nDataFrame after handling missing data:")
print(data.head())
```

Creating Dummy Variables

```python

data = pd.get_dummies(data, columns=['region_name'], drop_first=True)
print("\nDataFrame after creating dummy variables:")
print(data.head())
```

 Scaling Features


```python
scaler = MinMaxScaler()
data[['decile1', 'decile2', 'decile3']] = scaler.fit_transform(data[['decile1', 'decile2', 'decile3']])
print("\nDataFrame after scaling features:")
print(data.head())
```

Creating Interaction Terms


```python
data['decile1_times_decile2'] = data['decile1'] * data['decile2']
print("\nDataFrame after creating interaction terms:")
print(data.head())
```

Binning or Bucketing


```python
bins = [2000, 2010, 2020]
labels = ['2000-2010', '2010-2020']
data['reporting_year_category'] = pd.cut(data['reporting_year'], bins=bins, labels=labels)
print("\nDataFrame after binning 'reporting_year':")
print(data.head())
```


Convert 'reporting_year' to datetime format


```python
data['reporting_year'] = pd.to_datetime(data['reporting_year'])
data.head()
```

 Date Features


```python
data['reporting_month'] = data['reporting_year'].dt.month
data['reporting_year'] = data['reporting_year'].dt.year
print("\nDataFrame after extracting date features:")
print(data.head())
```

Log Transformations


```python
data['decile1_log'] = np.log1p(data['decile1'])
print("\nDataFrame after log transformation:")
print(data.head())
```

Feature Engineering Based on Domain Knowledge


```python
data['gdp_per_capita'] = data['reporting_gdp'] / data['reporting_pop']
print("\nDataFrame after adding 'gdp_per_capita' feature:")
print(data.head())
```

##Predictive Model

RandomForestClassifier

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
# Set the threshold for high and low reporting GDP
threshold_value = data['reporting_gdp'].mean()

# Create a binary target variable
data['high_reporting_gdp'] = (data['reporting_gdp'] > threshold_value).astype(int)

features = data[[
    'poverty_line', 'headcount', 'poverty_gap', 'gini', 'reporting_pop'
]]

# Define the target variable
target = data['high_reporting_gdp']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Standardize features if necessary
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and train the RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_rep)
```

Support Vector Machine (SVM) classifier

```python
from sklearn.svm import SVC

# Create and train the Support Vector Machine (SVM) classifier
svm_model = SVC(random_state=42)
svm_model.fit(X_train_scaled, y_train)

# Make predictions on the testing set
svm_y_pred = svm_model.predict(X_test_scaled)

# Evaluate the SVM model
svm_accuracy = accuracy_score(y_test, svm_y_pred)
svm_conf_matrix = confusion_matrix(y_test, svm_y_pred)
svm_classification_rep = classification_report(y_test, svm_y_pred)

print(f"SVM Model Accuracy: {svm_accuracy:.4f}")
print("\nSVM Model Confusion Matrix:")
print(svm_conf_matrix)
print("\nSVM Model Classification Report:")
print(svm_classification_rep)
```

## Apply data visualization acccording to the predictive models

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Calculate ROC curve and AUC for RandomForestClassifier
rf_prob = model.predict_proba(X_test_scaled)[:, 1]
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_prob)
rf_auc = roc_auc_score(y_test, rf_prob)

# Calculate ROC curve and AUC for SVM
svm_prob = svm_model.decision_function(X_test_scaled)
svm_fpr, svm_tpr, _ = roc_curve(y_test, svm_prob)
svm_auc = roc_auc_score(y_test, svm_prob)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(rf_fpr, rf_tpr, label=f'RandomForestClassifier (AUC = {rf_auc:.2f})', color='blue')
plt.plot(svm_fpr, svm_tpr, label=f'SVM Classifier (AUC = {svm_auc:.2f})', color='orange')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')

plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
